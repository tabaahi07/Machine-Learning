```python
# Linear Regression with multiple variable
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('price_of_house.csv')
df
```

```python
# As 1 value in bedroom is missing so first we need to handle it called data preprocessing , handling Nan values
# To fill that NA value , we fill it with median value
import math
med = math.floor(df.bedroom.median())
df.bedroom = df.bedroom.fillna(med)
df
```

```python
# Train model
reg = linear_model.LinearRegression()
reg.fit(df[['area' , 'bedroom' , 'age']] , df['price'])
```

```python
reg.coef_
reg.intercept_
reg.predict([[3000,3,40]])
```
