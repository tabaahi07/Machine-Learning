```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
import pandas as pd
df = pd.DataFrame(data.data)
df.columns = data.feature_names
df['target'] = data.target
df.head()
```

```python
X = df.drop('target' , axis=1)
from sklearn.model_selection import train_test_split
xtrain , xtest , ytrain , ytest = train_test_split(X,df['target'],train_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
logModel = LogisticRegression(solver='liblinear' , max_iter=50)
logModel.fit(xtrain,ytrain)
logModel.score(xtest,ytest)
```

```python
from sklearn.svm import SVC
svmModel = SVC(kernel='rbf' , C=10)
svmModel.fit(xtrain,ytrain)
svmModel.score(xtest,ytest)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfModel = RandomForestClassifier(criterion='entropy' , n_estimators=30)
rfModel.fit(xtrain,ytrain)
rfModel.score(xtest,ytest)
```

```python
from sklearn.model_selection import KFold , cross_val_score
kf = KFold(n_splits=4)
logistic_scores = cross_val_score(logModel, X, df['target'], cv=kf)
random_forest_scores = cross_val_score(rfModel, X, df['target'] , cv=kf)
svm_scores = cross_val_score(svmModel, X, df['target'] , cv=kf)
print(f"logistic score : {logistic_scores} , Random score score : {random_forest_scores} , svm score : {svm_scores} ")
```

```python
import numpy as np
model1 = np.mean(logistic_scores)
model2 = np.mean(random_forest_scores)
model3 = np.mean(svm_scores)
print(f"Log Reg Score : {model1} , RF Score : {model2} , SVM : {model3}")
```
