```python
import pandas as pd
df = pd.read_csv('hiring.csv')
df.head()
```

```python
df['experience'] = df['experience'].map({'one':1 , 'two':2 , 'three':3 , 'four':4 , 'five':5 , 'six':6 , 'seven':7,'eight':8,'nine':9,'ten':10,'eleven':11})
df
```

```python
X = df.drop('salary($)',axis=1)
X.head()
```

```python
X = X.fillna(0)
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,df['salary($)'])
```

```python
model.coef_
model.intercept_
```

```python
model.predict([[2,9,6] , [12,10,10]])
```
